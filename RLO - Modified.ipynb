```python
# Write to .dat file in AMPL format
with open("filtered_output_combine.dat", "w") as f:
    # Write the set
    f.write("set RAKELINKS := ")
    f.write("{} ".format(filtered_df_primary['ID_RAKE_LINK']))
    for val in df_option5['ID_RAKE_LINK']:
        f.write(f"{val} ")
    f.write(";\n\n")

    f.write("set SERVICES := ")
    f.write("{} ".format(filtered_df_primary['SERVICES(k)']))
    f.write("{} ".format(df_option5['SERVICES(2)'][0]))
    f.write("{} ".format(df_option5['SERVICES(k)'][0]))
    f.write("{} ".format(df_primary['SERVICES(k)'][1]))
    f.write(";\n\n")

    f.write("set STATIONS := ")
    f.write("{} ".format(filtered_df_primary['sta_dep']))
    f.write("{} ".format(filtered_df_primary['sta_arr']))
    f.write("{} ".format(df_option5['sta_arr(2)'][0]))
    f.write(";\n\n")
    
    f.write("param: T_a := ")
    f.write("{} ".format(filtered_df_primary['T_a']))
    f.write("{} ".format(df_option5['T_a(2)'][0]))
    f.write("{} ".format(df_option5['T_a(3)'][0]))
    f.write("{} ".format(df_primary['T_a'][1]))
    f.write(";\n\n")

    f.write("param: T_d := ")
    f.write("{} ".format(filtered_df_primary['T_d']))
    f.write("{} ".format(df_option5['T_d (2)'][0]))
    f.write("{} ".format(df_option5['T_d(3)'][0]))
    f.write("{} ".format(df_primary['T_d'][1]))
    f.write(";\n\n")

    f.write("param: T_h(MIN) := ")
    f.write("{} ".format(filtered_df_primary['T_h(MIN)']))
    f.write(";\n\n")

    f.write("param: T_clean := ")
    f.write("{} ".format(df_option5['T_clean'][0]))
    f.write(";\n\n")

    f.write("param: D(k) := ")
    f.write("{} ".format(filtered_df_primary['D(k) ']))
    f.write("{} ".format(df_option5['D(k) '][0]))
    f.write("{} ".format(df_option5['D(k) '][0]))
    f.write("{} ".format(df_primary['D(k) '][1]))
    f.write(";\n\n")

    
    f.write("param: T_prime := ")
    f.write("{} ".format(df_option5['T_prime'][0]))
    f.write(";\n\n")
   
    f.write("param: T_RJ(MIN) := ")
    f.write("{} ".format(filtered_df_primary['T_RJ(MIN)']))
    f.write("{} ".format(df_option5['T_RJ(MIN)'][0]))
    f.write("{} ".format(df_option5['T_RJ(MIN)'][0]))
    f.write("{} ".format(df_primary['T_RJ(MIN)'][1]))
    f.write(";\n\n")

    f.write("param: sta_arr := ")
    f.write("{} ".format(filtered_df_primary['sta_arr']))
    f.write("{} ".format(df_option5['sta_arr(2)'][0]))
    f.write("{} ".format(df_option5['sta_arr(3)'][0]))
    f.write("{} ".format(df_primary['sta_arr'][1]))
    f.write(";\n\n")

    f.write("param: sta_dep := ")
    f.write("{} ".format(filtered_df_primary['sta_dep']))
    f.write("{} ".format(df_option5['sta_dep(2)'][0]))
    f.write("{} ".format(df_option5['sta_dep(3)'][0]))
    f.write("{} ".format(df_primary['sta_dep'][1]))
    f.write(";\n\n")

    f.write("param: B := ")
    f.write("{} ".format(df_option5['B'][0]))
    f.write(";\n\n")

    f.write("param: M1 := ")
    f.write("{} ".format(df_option5['M1'][0]))
    f.write(";\n\n")

    f.write("param: M2 := ")
    f.write("{} ".format(df_option5['M2'][0]))
    f.write(";\n\n")

    f.write("param: sta_Origin := ")
    f.write("{} ".format(filtered_df_primary['sta_Origin']))
    f.write("{} ".format(df_option5['sta_Origin (#2)'][0]))
    f.write(";\n\n")

    f.write("param: T_Origin := ")
    f.write("{} ".format(filtered_df_primary['T_Origin']))
    f.write("{} ".format(df_option5['T_Origin(2)'][0]))
    f.write(";\n\n")

    f.write("param: e1 := ")
    f.write("{} ".format(df_option5['e1'][0]))
    f.write(";\n\n")

    f.write("param: TOTAL_COACHES(k) := ")
    f.write("{} ".format(filtered_df_primary['TOTAL_COACHES(k)']))
    f.write("{} ".format(df_option5['TOTAL_COACHES(k)'][0]))
    f.write("{} ".format(df_option5['TOTAL_COACHES(k)'][0]))
    f.write("{} ".format(df_primary['TOTAL_COACHES(k)'][1]))
    f.write(";\n\n")
    
  
print("filtered_output_Combine.dat file created sucessfully")

    

    
```
